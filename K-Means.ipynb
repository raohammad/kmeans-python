```python
df = pd.read_csv("/Users/hammadakhan/Documents/EISTI/DataMining/k-means/whiskies.txt")
df.drop(['RowID'], inplace=True, axis=1)
remove_tab = lambda x: x.replace('\t','')
df.Postcode = df.Postcode.map(remove_tab)
```

```python
X = df.drop(['Distillery','Postcode',' Latitude',' Longitude'], axis=1)
```

```python
X.head()
```

```python
#feature scaling
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

```python
# Finding K - below function is used
def find_k(X, k_range, sample_percent=1):
    #X: Data to cluster
    #k_range: range of K values
    from sklearn.cluster import KMeans
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.spatial.distance import cdist
    from sklearn.metrics import pairwise_distances
    
    N=X.shape[0]
    sampleSize = X.shape[0] * sample_percent
    
    if sampleSize > 0:
        index = np.arange(np.shape(X)[0])
        np.random.shuffle(index)
        X = X[index, :]
        
    mean_distortions = []
    for k in k_range:
        #cluster using K, then calculate the mean distortion (average distance to closest centroid)
        kmeans_model=KMeans(n_cluster=k, init='k-means++', n_jobs=-1).fit(X)
        mean_distortions.append(sum(np.min(pairwise_distances(X, kmeans_model.cluster_centers_,matrix='euclidean'),axis=1))/X.shape[0])
    
    mean_distortions
    #visualize results
    plt.plot(k_range, mean_distortions)
    plt.xlabel("K Value")
    plt.ylabel("Mean Distortion")
    plt.title("Elbow Graph for Mean Distortion per K")
    plt.show()    
```

```python
find_k(np.matrix(X), range(1-10), 1)
```

```python
kmeans_model = KMeans(n_clusters=4, random_state=4)
kmeans_model.fit(x_scaled)
```
